# Web Search function made

<h3>Shakespeare provides the web Search results about your query ( in shakespearen English)</h3>

```python

def Shakespeare_websearch():
    ddg_url = "https://duckduckgo8.p.rapidapi.com/"
    websearch = str(input("Enter your query to perform web search | Ent'r thy qu'ry to p'rf'rm web searcheth |   :"))
    ddg_querystring = {"q": websearch}
    ddg_headers = {
        "X-RapidAPI-Key": "f55c71370dmsh32bf0ad6b43a521p1a96ffjsn9d03d42f1873",
        "X-RapidAPI-Host": "duckduckgo8.p.rapidapi.com"
    }

    # Shakespeare Translator API
    translator_url = "https://shakespeare.p.rapidapi.com/shakespeare.json"
    translator_headers = {
        "X-FunTranslations-Api-Secret": "<REQUIRED>",
        "X-RapidAPI-Key": "f55c71370dmsh32bf0ad6b43a521p1a96ffjsn9d03d42f1873",
        "X-RapidAPI-Host": "shakespeare.p.rapidapi.com"
    }

    # Make the DuckDuckGo API request
    ddg_response = requests.get(ddg_url, headers=ddg_headers, params=ddg_querystring)
    ddg_data = ddg_response.json()

    if 'results' in ddg_data:
        results = ddg_data['results']

        for result in results:
            description = result.get('description', 'N/A')

            # Make the Shakespeare Translator API request
            translator_querystring = {"text": description}
            translator_response = requests.get(translator_url, headers=translator_headers, params=translator_querystring)
            translator_data = translator_response.json()

            if 'contents' in translator_data:
                translated_text = translator_data['contents'].get('translated', 'N/A')
                print("Original Text:", description)
                print("Translated Text:", translated_text)
                print("------------------")
            else:
                print("Translation not available for:", description)
                print("------------------")
    else:
        print("No search results found | Nay searcheth results hath found |.")

```

# News Article function made

<h3>Shakespeare gives top 20 news article about your query in shakespearen English.</h3>

```python

def Shakespeare_newssearch():


    query = str(input("Enter your query to find news articles | Ent'r thy qu'ry to findeth news articles |:    "))

    url = "https://contextualwebsearch-websearch-v1.p.rapidapi.com/api/search/NewsSearchAPI"

    querystring = {
        "q": query,
        "pageNumber": "1",
        "pageSize": "20",
        "autoCorrect": "true",
        "fromPublishedDate": "null",
        "toPublishedDate": "null"
    }

    headers = {
        "X-RapidAPI-Key": "f55c71370dmsh32bf0ad6b43a521p1a96ffjsn9d03d42f1873",
        "X-RapidAPI-Host": "contextualwebsearch-websearch-v1.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    data = response.json()

    if "_type" in data and data["_type"] == "news" and "value" in data:
        news_articles = data["value"]
        for article in news_articles:
            article_title = article.get("title", "")
            article_description = article.get("description", "")

            translation_url = "https://shakespeare.p.rapidapi.com/shakespeare.json"

            translation_querystring = {"text": article_description}

            translation_headers = {
                "X-FunTranslations-Api-Secret": "<REQUIRED>",
                "X-RapidAPI-Key": "f55c71370dmsh32bf0ad6b43a521p1a96ffjsn9d03d42f1873",
                "X-RapidAPI-Host": "shakespeare.p.rapidapi.com"
            }

            translation_response = requests.get(translation_url, headers=translation_headers, params=translation_querystring)

            translation_data = translation_response.json()

            if "contents" in translation_data and "translated" in translation_data["contents"]:
                translated_text = translation_data["contents"]["translated"]
                print("Title:", article_title)
                print("Translated Description:", translated_text)
                print()
            else:
                print("Translation not available for article:", article_title)
    else:
        print("No news articles found.")


```

# Function for Shakespeare Information Retrieval System

<h3>Use the Shakespeare Info Retrieval functionality to get the contents of a webpage in Old English.<h3/>

```python


# Function to translate text to Shakespearean language
def translate_to_shakespeare(text):
    url = "https://shakespeare.p.rapidapi.com/shakespeare.json"
    querystring = {"text": str(text)}
    headers = {
        "X-FunTranslations-Api-Secret": "<REQUIRED>",
        "X-RapidAPI-Key": "f55c71370dmsh32bf0ad6b43a521p1a96ffjsn9d03d42f1873",
        "X-RapidAPI-Host": "shakespeare.p.rapidapi.com"
    }
    response = requests.get(url, headers=headers, params=querystring)
    return response.json().get('contents', {}).get('translated')

# Function to translate paragraphs of a web page to Shakespearean language
def translate_webpage_to_shakespeare():
    # Get input URL from the user
    input_url = input("Enter the URL of the webpage: ")

    # Send a GET request to the URL
    response = requests.get(input_url)

    # Extract contents from the webpage
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract and print <p> tags
        p_tags = soup.find_all('p')
        for p in p_tags:
            original_text = p.text.strip()
            translated_text = translate_to_shakespeare(original_text)
            print("Original Text: ", original_text)
            print("Shakespearen Text: ", translated_text)
            print("***************************************************************************************")

    else:
        print("Failed to retrieve the webpage.")

# Call the function to translate the web page
```
